# **Example 0:  download data for our games with fixed submission id**


Algorithm has 3 steps:
1. Get info with played episodes, teams info and teams submissions using curl request.
2. Parse data and save information to files. If information is already exists script will add only new data to existing files.
3. The base data is ready, than we download replays (curl request) for new games and create extended dataframe with different params for furute analysing.

You can separate submissions  data in different files using param `player_name` . If param is default all requests will be concantenated and written to 1 file.

**our strong games**

```python
all_df = get_episodes_extended('my')

print("STRONG")
for my, my_name in [(my_a, 'my_a'), (my_0, 'my_0'), (my_1, 'my_1'), (my_2, 'my_2')]:
    for frog, frog_name in [(frog_1, 'frog_1'), (frog_2, 'frog_2'), (frog_old_1, 'frog_old_1'), (frog_old_2, 'frog_old_2'), 
                            (adbg_1, 'adbg1'), (adbg_2, 'adbg2'),
                              (oneup_1, 'oneup1'), (gregor_1, 'gregor_1'), (gregor_2, 'gregor_2')]:
        df = all_df[(all_df['is_strong'] == 1) & (all_df['main_submission_id'] == my) & (all_df['enemy_submission_id'] == frog)]
        my_win = df['main_reward'].sum()
        enemy_win = df['enemy_reward'].sum()
        sum_matches = my_win + enemy_win
        my_winrate = my_win / max(1, sum_matches) * 100
        enemy_winrate = enemy_win / max(1, sum_matches) * 100
        if sum_matches > 0:
            print(f'ENEMY Winrate Points {my_name:10} vs {frog_name:11}: {enemy_win:3} of {sum_matches:3} = {enemy_winrate:.1f}%')
```
